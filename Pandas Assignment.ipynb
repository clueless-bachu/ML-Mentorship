```python
pclass = [1,2,3]
embarked = ['S','Q','C']
for i in pclass:
    print('Avg cost and std for class {}:'.format(i),\
    train_df[train_df["Pclass"]==i]['Fare'].mean(),'and',train_df[train_df["Pclass"]==i]['Fare'].std())
print('\n')
for i in embarked:
    print('Avg cost and std for Embark Location {}:'.format(i),\
    train_df[train_df["Embarked"]==i]['Fare'].mean(),'and',train_df[train_df["Embarked"]==i]['Fare'].std())
```
